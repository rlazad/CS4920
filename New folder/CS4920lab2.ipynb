```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
#--------------------------------------#
path = os.getcwd()
files = os.listdir(path)
files
```

```python
files_csv = [f for f in files if f[-3:] == 'csv']
files_csv
```

```python
list_csv = []
for f in files_csv:
   df = pd.read_csv(f, index_col=None, header=0)
   list_csv.append(df)
    
frame = pd.concat(list_csv, axis=0, ignore_index=True)
```

```python
frame.replace([np.inf, -np.inf, np.nan, 'Nan', -1], 0)
```
